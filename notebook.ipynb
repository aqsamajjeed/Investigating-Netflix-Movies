**Importing libraries and loading dataset**

```python
# Importing pandas and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

**What was the most frequent movie duration in the 1990s?**

```python
#Filter for movies only
is_movie=netflix_df.loc[:,"type"]=="Movie"
netflix_df[is_movie]

#filter for year 1990s
yearr=np.logical_and(netflix_df.loc[:,"release_year"]>1989,netflix_df.loc[:,"release_year"]<2000)
netflix_df[yearr]

#filter for movies released in 1990s
movieWyear=np.logical_and(is_movie,yearr)
filtered=(netflix_df[movieWyear])

#ploting the distribution
plt.hist(filtered["duration"],bins=20)
plt.title("Distribution of Movies in 1990s")
plt.xlabel("Duration(movies)")
plt.ylabel("Frequency")
plt.show()
plt.clf()

#most frequent duration of movies
duration=100
print(f"The most frequent duration of movies is {duration} minutes.")
```

**Count the number of short action movies released in the 1990s**

```python
#filtering genre of action movies
action_movies=(netflix_df.loc[:,"genre"]=="Action") & movieWyear
action_movies_df=netflix_df[action_movies]

#counting number of action movies
short_movie_count=0
for label,rows in action_movies_df.iterrows():
    if rows["duration"]<90:
        short_movie_count=short_movie_count+1
print("The number of short action movies released in 1990s is " + str(short_movie_count)) 
```
